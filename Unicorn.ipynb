```python
data.head()
```

```python
df1 = data.sort_values('ROI' , ascending=False)
```

```python
df1.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df1["Company"][:10],df1["ROI"][:10])
f={"size":15,"color":"blue","family":"Times New Roman"}
f1={"size":15,"color":"black","family":"Times New Roman"}
plt.xlabel("Unicorn Name",fontdict=f)
plt.ylabel("Return on Investment",fontdict=f)
plt.title("Unicorn Companies having biggest return",fontdict=f1)
plt.xticks(rotation=60)
plt.rcParams["figure.figsize"]=(7,5)
plt.show()
```
